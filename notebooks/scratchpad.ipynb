```python
def plot_trade_regional(df, region_list, colors):
    fig = plt.figure(figsize=(20,10))
    ax = fig.add_subplot(1, 1, 1)
    
    x_data = range(1960, 2021)
    
    for i, reg in enumerate(region_list):
        ax.plot(x_data, df.loc[reg], label=reg, color=colors[i])
    
    ax.legend()
    ax.set_title('Exports as % of GDP by Region')
    ax.set_xlabel('Year')
    ax.set_ylabel('Percent')
```

```python
colors = ['green', 'blue', 'red', 'gray', 'black', 'orange', 'purple']
plot_trade_regional(exports_df2, region_list, colors)
```
